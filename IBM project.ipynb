```python
import seaborn as sns
sns.boxplot(df1['Age'])
```

```python
numerical_columns= df1.select_dtypes(include=['number']).columns.tolist()
numerical_columns
```

```python
categorical_columns= df1.select_dtypes(include=['object']).columns.tolist()
categorical_columns
```

```python
import matplotlib.pyplot as plt
```

```python
fig, axes = plt.subplots(nrows=len(numerical_columns), ncols=1, figsize=(10, 6*len(numerical_columns)))

for i, col in enumerate(numerical_columns):
    sns.boxplot(x=df1[col], ax=axes[i])  
    
    z_score = stats.zscore(df1[col])
    threshold = 3
    outliers_indices = (z_score > threshold) | (z_score < -threshold)
    df2 = df1[~outliers_indices]

plt.tight_layout()  
plt.show()
```
